```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel (r'Car_Delarship_Dataset.xlsx')
df
```

<font color='Green' size=5>Data Pre-processing</font>

<font color='Green' size=2>Checking duplicate records in dataset</font>

```python
duplicateRowsDF = df[df.duplicated(['user_id', 'entry_method', 'dealership_id', 'new_status_str', 'prior_status_str', 'created_at', 'triggering_user_id'])]
duplicateRowsDF
```

<font color='Green' size=3>Drop duplicate data from the dataset</font>

```python
df = df.drop_duplicates(subset=['user_id', 'entry_method', 'dealership_id', 'new_status_str', 
                                'prior_status_str', 'created_at', 'triggering_user_id'])
```

<font color='Green' size=3>Convert Date</font>

```python
df['created_at'] =  pd.to_datetime(df['created_at']).dt.date
# df['date'] = pd.to_datetime(df['date'])
df.sort_values(by='created_at')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.dropna()
df
```

```python
df["entry_method"] = df["entry_method"].astype('category')
df.dtypes
```

```python
df["entry_method"] = df["entry_method"].cat.codes
df.tail(11)
```

```python
df.describe()
```

<font color='Green' size=3>Find unique value from dealership_id column</font>

```python
df_dealership_id_list = list(df['dealership_id'].unique())
print(df_dealership_id_list)
```

<font color='Green' size=4>Analysis #1</font>

<font color='Green' size=3>Code to count user_id for each dealer_id</font>

```python
for i in range(len(df_dealership_id_list)):
    new_df = df[df["dealership_id"] == df_dealership_id_list[i]]
    unique_user_id = new_df["user_id"].unique()
    for j in range(len(unique_user_id)):
        usr_df = new_df[new_df["user_id"] == unique_user_id[j]]
        cnt = usr_df.groupby('created_at').size().rename('Count')
        if j == 0:
            result = usr_df.drop_duplicates(subset='created_at').merge(cnt, left_on='created_at', right_index=True)
        else:
            new_result = usr_df.drop_duplicates(subset='created_at').merge(cnt, left_on='created_at', right_index=True) 
            result = result.append(new_result, ignore_index=True)
    if i == 0:
        new_dataset = result.copy()
    else:
        new_dataset = new_dataset.append(result,ignore_index=True)   
```

```python
new_datasets = new_dataset[['user_id', 'dealership_id', 'created_at', 'Count']]
new_datasets
```

<font color='Green' size=3>Plot dealer_ship and user_id relationship</font>

```python
plt.tight_layout()
g = sns.FacetGrid(new_dataset, col="dealership_id", hue="user_id", height=3, aspect=3, col_wrap=3, sharex = False, sharey = False)
g.map_dataframe(sns.scatterplot, x="created_at", y="Count")
g.set_axis_labels("DATE_CREATED", "Count")
# g.add_legend()
plt.savefig("dealer_ship-user_id--Relationship.jpg")
```

<font color='Green' size=4>Analysis #2</font>

```python
df_created_at = df["created_at"].unique()
df_created_at
```

<font color='Green' size=3>Code to count entry_method for each dealer_id for a particular date</font>

```python
for i in range(len(df_created_at)):
    new_df = df[df["created_at"] == df_created_at[i]]
    dealer_array = new_df["dealership_id"].unique()
    for j in range(len(dealer_array)):
        new_data = new_df[new_df['dealership_id'] == dealer_array[j]]
        cnt = new_data.groupby('entry_method').size().rename('Count')
        if j == 0:
            result = new_data.drop_duplicates(subset='entry_method').merge(cnt, left_on='entry_method', right_index=True)
        else:
            new_result = new_data.drop_duplicates(subset='entry_method').merge(cnt, left_on='entry_method', right_index=True) 
            result = result.append(new_result, ignore_index=True)
    if i == 0:
        new_datasets_2 = result
    else:
        new_datasets_2 = new_datasets_2.append(result,ignore_index=True)
```

```python
new_datasets_2 = new_datasets_2[['user_id', 'entry_method', 'dealership_id', 'created_at', 'Count']]
new_datasets_2
```

<font color='Green' size=3>Plot dealer_ship and entry_method relationship</font>

```python
cat_array = [0,1]
for i in range(len(cat_array)):
    new_df = new_datasets_2[new_datasets_2['entry_method'] == cat_array[i]]
    dealer_array=new_df["dealership_id"].unique()
    if len(dealer_array) > 1:
        plt.figure(figsize=(15,8))
        for j in range(len(dealer_array)):
            df1 = new_datasets_2[new_datasets_2["entry_method"]== cat_array[i]]
            df1 = df1[df1["dealership_id"] == dealer_array[j]]
            if len(df1) > 0:
                plt.bar(df1['created_at'],df1['Count'],label=dealer_array[j])
                plt.xticks(rotation=90)
        plt.legend()
        if(i == 0):
            title="Dashboard"
            plt.savefig("dealer_ship-Dashboard-Usage.jpg")
        else:
            title="QR code"
            plt.savefig("dealer_ship-QR_code-Usage.jpg")
        plt.title(title)
        plt.show()
```

<font color='Green' size=4>Analysis #3</font>

<font color='Green' size=3>Convert date-time</font>

```python
df['created_at'] =  pd.to_datetime(df['created_at']).dt.round('H')
df
```

```python
series_data = df['created_at'].value_counts().sort_index()
plt.figure(figsize=(20,10))

plt.yticks(rotation=90)
# series_data.plot(kind="bar")
plt.xticks(rotation=90, ha='right')
plt.plot(series_data)
plt.show()
plt.savefig("peak_anaylis.jpg")
```
